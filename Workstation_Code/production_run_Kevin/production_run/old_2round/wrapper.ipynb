```python
def subtract_atom(molecule, index, bounds):
    for i in range(len(molecule)):
        molecule[i][index] = round(float(molecule[i][index]) - bounds, 3)
        
    return molecule

def add_atom(molecule, index, bounds):
    for i in range(len(molecule)):
        molecule[i][index] = round(float(molecule[i][index]) + bounds, 3)
        
    return molecule
```
