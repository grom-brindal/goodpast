```python
for i in range(len(molecules)):
    atom_to_be_written = []
    molecule_to_be_written = []
    for x in range(6):

        atom = molecules[i][x].split()
        atom_to_be_written.append([atom[2], atom[5], atom[6], atom[7]])
    molecule_to_be_written.append(atom_to_be_written)
    print(molecule_to_be_written)


    cas_input_file_name = '.\\cas_input_files\\' + str(i) + '.in'
    with open(cas_input_file_name, 'w') as f:
        f.write('!leave this line blank\n\n')
        f.write('gprint, orbitals=30\n')
        f.write('memory, 125., m\n')
        f.write('symmetry, nosym\n')
        f.write('nelect=16\n')
        f.write('basis=6-31g*\n\n')
        f.write('geometry={\n')
        for r in range(len(molecule_to_be_written[0])):
            atom_to_be_written = molecule_to_be_written[0][r]
            if i == 0:
                print(atom_to_be_written)
            if i == len(molecules) - 1:
                print(atom_to_be_written)
            f.write(str(atom_to_be_written[0]) + ' ' + str(atom_to_be_written[1]) + ' ' + str(atom_to_be_written[2]) + ' ' + str(atom_to_be_written[3]) + '\n')
        
        f.write('}\n\n')
        f.write('hf\n')
        f.write('{multi,\n')
        f.write('closed 7\n')
        f.write('occ 9\n')
        f.write('wf,spin=0;STATE,3\n')
        f.write('}\n\n')


print(molecule_to_be_written)
```
